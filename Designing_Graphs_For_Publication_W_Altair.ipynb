## Why use the Python Pathlib library?

File paths and pathlib

You may or may not have seen that macs and windows have different formats for file paths. This means that if you are trying to programmatically generate a path that will work on multiple operating systems, you have to do some strange things.

Also, even if you are not worring about that, trying to just get the file name out of an arbitrarily deep file path usually would result in a ton of strange string manipulations. Also, if you were trying to get the root name of a file out so you could add a different file extension, you would have to do some even worse manipulation.

This is where pathlib comes in. It understands the file system from an object oriented perspective, so you are operating on and extracting information from the file path using object attributes and methods rather than manipulating a string without conceptuale awareness of the core content meaning.
First, import pathlib. This will give you access to the Path() function. You can pass this function a full file name or a directory name. Passing it a directory will give you access to the glob() method for searching for file paths.

## Pathlib and Labs

Theoretically, if you have data in various different directories you can use pathlib to open them while not needing the jupyternotebook to exists in that directory to access the data. When you share your research data you can thus maintain your file organization while the user can run the cells.

### Exact Graph Replicated 

```python
aminoAcids = ['NN-SS 1.5mM-F02', 'NN-TT 1.5mM-F04', 'NN-VV 1.5mM-F05', 'NN-AA 1.5mM-F08']

Figure06A= alt.Chart(Figure6A, title='a)').mark_line().encode(
    alt.X("Wavelength:Q", title="Wavelength / nm"), # , scale=alt.Scale(domain=[400, 660])
    alt.Y("Flourescence_x10\u00b3:Q", title="Fluorescence (x10\u00b3) /a.u.", scale=alt.Scale(domain=[0, 5])),
    alt.Color('AminoAcidSideChains:N', title="Amino Acid Side Chains"),
    tooltip=['AminoAcidSideChains']
).transform_filter(
    alt.FieldOneOfPredicate(field='AminoAcidSideChains', oneOf=aminoAcids)
).configure(
    font='Arial' #Default font for all text marks, titles, and labels.
).configure_title(
    anchor='start',
    fontSize=12
).configure_legend(
    orient='top-right'
).configure_axis(
    grid=False,
    labelFontSize=12,
    titleFontSize=12
).configure_view(
    strokeWidth=0
)


Figure06A # you can save the graph as a PNG by clicking the ellipses in the top-rigth corner
```

##  Ways of creating templates that can make plots consistent 



- [Customizing Visualizations](https://altair-viz.github.io/user_guide/customization.html?highlight=theme#customizing-visualizations)
    - This section outlines some of the ways users can make adjustments to visualizations
    - for example: setting themes, adjusting axis limits (e.g., scale/domain), adjust axis labels (e.g., format, titles, ), adjust legends, removing chart border, customize colors, Adjusting the width of Bar Marks and chart, adjusting chart size
- [Altair Documentaiton: Top-level-Chart-Configuration](https://altair-viz.github.io/user_guide/configuration.html#top-level-chart-configuration)
    - Think of the configure_*() methods as setting a chart theme (i.e., set the default styles for the entire chart)
- [Altair Themes](https://altair-viz.github.io/user_guide/configuration.html#altair-themes)
    - A theme, in altair, is a set of chart configurations applied globally each python session. This means you can **produce similar-looking visualizations consistently**
    - You create a theme and then apply chart configurations globally within any Python session
    - done using the alt.theme object
    - Since Altair uses Vega in their backend programing you can use any of the preconstructed themes from [Vega Themes](https://github.com/vega/vega-themes/#included-themes). If you find one that is close to what you want, you can copy the code and adjust as fit. Read more on making a a theme here: [define a custom theme](https://altair-viz.github.io/user_guide/configuration.html#defining-a-custom-theme)
- [Consistently Beautiful Visualizations with Altair Themes](https://towardsdatascience.com/consistently-beautiful-visualizations-with-altair-themes-c7f9f889602)
    - A blog focused on using themes and how you can code one into your script
- [established pattern for importing Vega themes into Altair](https://github.com/altair-viz/altair/issues/1333)
    - a Github tread discussing how to import Themes 
    - the most consistent way is writing `alt.themes.enable('string of theme name goes here')`
- [Example Code of how to construct an Altair Chart Theme](https://pmbaumgartner.github.io/streamlitopedia/theming-altair.html)
    - code examples of someone who has used Altair to make themes.
    - it's a collection of free to use of tools, code snippets, and ideas that they found helpful when making [streamlit](https://streamlit.io/) apps. Even discuss markup language which is used in jupyter notbeook

```python
# example of using a pre-constructed theme in your visual 
# Using the with statement so you dont apply the theme to all charts but only the one in that cell
# If you would like to use any theme just for a single chart, you can use the `with` statement to enable a temporary theme:

with alt.themes.enable('ggplot2'):
    Figure06A= alt.Chart(Figure6A, title='a)').mark_line().encode(
    alt.X("Wavelength:Q", title="Wavelength / nm"),
    alt.Y("Flourescence_x10\u00b3:Q", title="Fluorescence (x10\u00b3) /a.u."),
    alt.Color('AminoAcidSideChains:N', title="Amino Acid Side Chains"),
    tooltip=['AminoAcidSideChains']
    ).transform_filter(
        alt.FieldOneOfPredicate(field='AminoAcidSideChains', oneOf=aminoAcids)
    )
    # Figure06A.show() # This command will block the Python interpreter until the browser window containing the chart is closed.
    Figure06A.display()
    # Note: that calling .display() on multiple charts is how you can display multiple charts in a single cell / image
    # more infor here: 
```
